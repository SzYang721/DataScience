## Q1. Sampling distributions using U(0,1)

Here we will understand how to sample from some common distributions. It is usually easy to sample from a uniform distribution using any of the multiple pseudorandom number algorithms that exist. However, suppose we are given some distribution $p(y)$ with an analytical form (e.g. Gaussian, exponential distribution). What would be the process to sample from such a distribution? There is a simple two step process involving the cumulative distribution function (CDF):

- 1. Calculate the CDF for $p$ by $$h(y) = \int_{-\infty}^y p(\hat{y}) d\hat{y}.$$
- 2. Use a uniform distribution $U(z \mid 0, 1)$ to transform the samples by $$y = h^{-1}(z).$$

Discuss with your partner or tutor to derive this method and understand why this works.

### Q1b. Sampling the exponential distribution.
Let's put the above theory to use to sample the exponential function 
$$
    p(y) = 
    \begin{cases}
        \lambda e^{-\lambda y}, & y \geq 0 \\
        0, & y < 0
    \end{cases}
$$
for with *rate parameter* $\lambda > 0$. Please prove that
- 1. The CDF of $p$ is $h(y) = 1 - e^{-\lambda y}$ and 
- 2. To sample using uniform samples $z$, we have $y = -\frac{1}{\lambda} \ln (1-z).$

Now try implementing the routine below for sampling the exponential function.

````python
# Solution
def sample_exponential(l):
    """ Return a sample from the pdf $p$ defined above. """
    
    z = np.random.uniform()
    y = -(1 / l) * np.log(1 - z)
    
    return y
````

## Q2. Rejection sampling
Sometimes the method above is difficult to implement or even impossible for example when $p$ does not have an analytic form. More generally, what do we do $p$ is difficult to directly sample from? Rejection sampling solves this issue with the help of of a proposal distribution $q$ which is much easier to sample from and satisfies the property for some constant $k>0$ and for all $z$ that
$$k\cdot q(z) \geq p(z).$$

The sampling method is as follows:
1. Generate a random number $z_0$ from the distribution $q(z)$
2. Generate a random number $u_0$ from the uniform distribution $U([0, kq(z_0)])$
3. If $u_0 > p(z_0)$, then reject the pair $(z_0, u_0)$

The remaining $(z_0, u_0)$ pairs have uniform distribution under the curve $p(z)$ while the $z_0$ values are distributed according to $p(z)$.

*Note*: the method described here is slightly different to the one described in lecture slides but it is easy to show that are equivalent. As an exercise, you can verify that they are indeed equivalent.

### Q2a. Implement rejection sampling
In this question we will implement rejection sampling to sample from the [gamma distribution](https://en.wikipedia.org/wiki/Gamma_distribution#Characterization_using_shape_%CE%B1_and_rate_%CE%B2)
$$p(z \mid a, b) = \frac{b^az^{a-1}\exp(-bz)}{\Gamma(a)}.$$
Note that the gamma distribution does not have an analytical form so rejection sampling is ideal for sampling. We will use a simple Gaussian as our proposal distribution. Fill in the missing code for our rejection sampling implementation below.

*Hint*: don't forget about $k$ in your implementation.

````python
# Solution
def p(z, alpha, beta):
    """Gamma pdf"""
    return scipy.stats.gamma.pdf(z, a=alpha, scale=1/beta)

def q(z, mu, sigma):
    """Gaussian pdf"""
    return scipy.stats.norm.pdf(z, mu, sigma)

def rejection_samples(k, mu, sigma, alpha, beta, num_samples=10000):
    """Return accepted and rejected pairs generated by rejection samples."""
    
    acc_samples_x = []
    acc_samples_y = []
    rej_samples_x = []
    rej_samples_y = []
    
    for _ in range(num_samples):
        
        # Step 1: generate z from q(z)
        z = np.random.normal(mu, sigma)
        
        # Step 2: generate u from U[0, k*q(z)]
        qz = q(z, mu, sigma)
        u = np.random.uniform(0, k * qz)
        
        # Step 3: accept (z, u) if u < p(z), otherwise reject.
        pz = p(z, alpha, beta)
        if u <= pz:
            acc_samples_x.append(z)
            acc_samples_y.append(u)
        else:
            rej_samples_x.append(z)
            rej_samples_y.append(u)
    
    return acc_samples_x, acc_samples_y, rej_samples_x, rej_samples_y
````

## Q3. Importance sampling
The goal of importance sampling is to estimate the expectation of some complicated probability distribution without needing to sample from such probability distribution. More specifically, we are interested in estimating for some distribution $p$ and function $f$ the value $\mathbb{E}_p[f]$. Similarly to rejection sampling, we will use the help of a proposal distribution $q$ that is easier to sample from. In this question, we will implement and understand importance sampling for an example $f$ and $p$. For a refresher on the details of importance sampling, see Bishop Sec. 11.1.4.

Let us define the function 2-dimensional $f(x, y)$ by
$$
  f(x, y) = x  y  \cos(x) \cos(y),
$$

and the (unnormalised!) distribution $\widetilde{p}(x, y)$ by
$$
  \widetilde{p}(x, y) = 
                 \exp \left\{- \frac{1}{4}((x - 2)^2 + (y - 3)^2) \right\}
    - \frac{3}{4} \exp \left\{- \frac{1}{2} ((x - 2)^2 + (y - 3)^2) \right\}.
$$

Our goal is to numerically estimate $\mathbb{E}_{p(x, y)}[f(x, y)]$ defined by
$$
   \mathbb{E}_{p(x, y)}[f(x, y)] = \int_{-\infty}^{\infty} f(x, y) \, p(x, y) \mathrm{d} x \mathrm{d} y
$$

where $ p(x, y) $ is the normalised probability distribution proportional to $\widetilde{p}(x, y)$, so
$$
   p(x, y) = \frac{ \widetilde{p}(x, y) }
                  { \int_{-\infty}^{\infty}  \int_{-\infty}^{\infty} \widetilde{p}(x, y) \mathrm{d} x \mathrm{d} y }.
$$

Note that with importance sampling, this can be achieved without calculating the normalisation. All we need to do, is to use sample points from an appropriate distribution $\widetilde{q}(x, y)$ 
which has most of its probability mass in regions where $\widetilde{p}(x, y)$ is also nonzero.

We will implement importance sampling to estimate $\mathbb{E}_{p(x, y)}[f(x, y)]$ for two choices of $\widetilde{q}(x, y)$, a uniform distribution and a Gaussian distribution. To help you get started, below are implementations for $f$, $\widetilde{p}$ and their plots.

````python
def f(x, y):
    return x * y * math.cos(x) * math.cos(y)

def p_tilde(x, y):
    s1 = 2.0
    s2 = 1.0
    p1 = math.exp(- 1.0/(2.0*s1) * ((x - 2)**2 + (y - 3)**2))
    p2 = math.exp(- 1.0/(2.0*s2) * ((x - 2)**2 + (y - 3)**2))
    return p1-3/4*p2
    
xs = np.linspace(-8,8,256)
ys = np.linspace(-8,8,256)
matf = np.array([[f(x, y) for x in xs] for y in ys])
matp = np.array([[p_tilde(x, y) for x in xs] for y in ys])
plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
plt.title('Heatmap of f')
plt.imshow(matf, interpolation='nearest', origin='lower', extent=[-8, 8, -8, 8], 
           vmax=abs(matf.flatten()).max(), vmin=-abs(matf.flatten()).max())
plt.subplot(1,2,2)
plt.title('Heatmap of p_tilde')
plt.imshow(matp, interpolation='nearest', origin='lower', extent=[-8, 8, -8, 8], 
           vmax=abs(matp.flatten()).max(), vmin=-abs(matp.flatten()).max())
plt.show()
````

### Q3a. Importance sampling using the uniform distribution
Complete the below implementation for estimating $\mathbf{E}_{p(x, y)}[f(x, y)]$ with the uniform proposal distribution $\widetilde{q}(x, y) = [-10, 20]\times[-10, 20]$ for the input number of samples ```num_samples```.

Hint: look at Bishop Eqn. 11.22. Note that the denominator in Eqn. 11.23 is constant for all $l$.

````python
# Solution
def importance_uniform(num_samples):
    """ Compute the expectation of f(x,y) over the unnormalised distribution p(x,y).
    Uses the uniform(-10,30) as a proposal distribution.
    """
    exp_f = 0
    denom = 0
    rv_x = uniform(loc=-10, scale=30)
    rv_y = uniform(loc=-10, scale=30)
    for _ in range(num_samples):
        x = rv_x.rvs()
        y = rv_y.rvs()
        pval = p_tilde(x, y)
        qval = rv_x.pdf(x)*rv_y.pdf(y)
        
        exp_f  += f(x, y) * pval/qval
        denom += pval/qval 
    return exp_f/denom
````

````python
# Use this cell to test your implementation.
# Depending on your implementation it might take a few seconds to run
print(f'Estimated expectation: {importance_uniform(50000):.3}')
print('True expectation: 0.671')
````

### Q3b. Importance sampling using the uniform distribution
Now complete the below implementation for estimating 
$\mathbb{E}_{p(x, y)}[f(x, y)]$ with the Gaussian proposal distribution 
$$\widetilde{q}(x, y) = {\mathscr{N}\left({(x, y)^T} ~\middle|~ {\mathbf{\mu} = (2, 3)^T}, {\mathbf{\Sigma} = \begin{bmatrix} 2 & 0 \\ 0 & 3 \\ \end{bmatrix}}\right)}.$$

Hint: the ```importance_gaussian``` implementation should be the same as your ```importance_uniform``` implementation above with the only difference in the sample generation.

````python
# Solution
def importance_gaussian(num_samples):
    """Compute the expectation of f(x,y) over the unnormalised distribution p(x,y).
    Uses Normal([2,3], [[2,0][[0,3]]]) as a proposal distribution.
    """
    exp_f = 0
    denom = 0
    rv = multivariate_normal(mean=np.array([2,3]), cov=np.array([[2,0],[0,3]]))
    for _ in range(num_samples):
        x, y = rv.rvs()
        pval = p_tilde(x, y)
        qval = rv.pdf([x,y])
        
        exp_f  += f(x, y) * pval/qval
        denom += pval/qval 
    return exp_f/denom
````

````python
# Use this cell to test your implementation. 
# Depending on your implementation it might take a few seconds to run
print(f'Estimated expectation: {importance_gaussian(50000):.3}')
print('True expectation: 0.671')
````

### Comparing the empirical and analytic results

For the function $f(x, y)$ and the normalised distribution $p(x, y)$,
the correct result can be computed analytically
$$
    \mathbf{E}_{p(x, y)}[f(x, y)]
     = \int_{-\infty}^{\infty} \int_{-\infty}^{\infty} f(x, y) \, p(x, y) 
                \mathrm{d} x \mathrm{d} y \\
     =   \frac{16 (5 \cos(1) + \cos(5) + \sin(1) - 5 \sin(5))
      - 3 e (7 \cos(1) + 5 \cos(5) + \sin(1) - 5 \sin(5)) }
               {10 \, e^2} \\                      
     \approx 0.670859
$$

Below we compare the convergence rate of the two approaches given above by plotting the two curves showing the empirical expectation as a function of the number of samples as well as the analytical value. Note that we have rewritten the functions above to make plotting a lot faster, though even then it will still take a few seconds to run.

````python
def est_uniform(batch=100, batch_size=1000):
    exp_f = 0
    denom = 0
    N = batch * batch_size
    results = np.empty((batch,))
    k = 0
    rv_x = uniform(loc=-10, scale=30)
    rv_y = uniform(loc=-10, scale=30)
    for ix in range(N):
        x = rv_x.rvs()
        y = rv_y.rvs()
        pval = p_tilde(x, y)
        qval = rv_x.pdf(x)*rv_y.pdf(y)
        exp_f  += f(x, y) * pval/qval
        denom += pval/qval 
        if ix % batch_size == 0:
            results[k] = exp_f/denom
            k += 1
    return results
def est_gaussian(batch=100, batch_size=1000):
    exp_f = 0
    denom = 0
    N = batch * batch_size
    results = np.empty((batch,))
    k = 0
    rv = multivariate_normal(mean=np.array([2,3]), cov=np.array([[2,0],[0,3]]))
    for ix in range(N):
        x,y = rv.rvs()
        pval = p_tilde(x, y)
        qval = rv.pdf([x,y])
        exp_f  += f(x, y) * pval/qval
        denom += pval/qval 
        if ix % batch_size == 0:
            results[k] = exp_f/denom
            k += 1
    return results
e_uniform = est_uniform()
e_gaussian = est_gaussian()
num_points = len(e_uniform)
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(111)
ax.plot(range(num_points), e_uniform, label='Uniform')
ax.plot(range(num_points), e_gaussian, label='Gaussian')
ax.plot([0, num_points], [0.670859, 0.670859], 'r--', label='analytic')
ax.legend(loc='lower right')
ax.set_ylim([0.670859-0.5, 0.670859+0.5]);
````
